```python
%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Make fake dataset
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
 
# Choose the position of each barplots on the x-axis (space=1,4,3,1)
y_pos = [0,0.75,1.55,2.25,3]
 
# Create bars
plt.bar(y_pos, height, edgecolor='black')
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
 
# Show graphic
plt.show()
```
