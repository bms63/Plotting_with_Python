# 2 Research Question

How has the percentage of hate crimes relative to population changed for Ann Arbor, Clinton Township, Detroit, Grand Rapids and Lansing from 2005 to 2016?


# 3 Links
You must provide at least two links to publicly accessible datasets. These could be links to
files such as CSV or Excel files, or links to websites which might have data in tabular form,
such as Wikipedia pages.

* https://ucr.fbi.gov/ - FBI home for Hate Crime Statistics
* https://ucr.fbi.gov/hate-crime - 2016-2005 Hate Crime Statistics in Excel Files

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [13, 8]



df = pd.read_csv('HateCrimes.csv')
df1 = df.T
df1.columns = df1.iloc[0]
df2 = df1.iloc[1:]
df2
```

```python


# Create plots with pre-defined labels.
fig, ax = plt.subplots()
ax.plot(df2['Ann Arbor'], label='Ann Arbor')
ax.plot(df2['Clinton Township'], label='Clinton Township')
ax.plot(df2["Detroit"], label='Detroit')
ax.plot(df2["Grand Rapids"], label='Grand Rapids')
ax.plot(df2["Lansing"], label='Lansing')

ax.annotate(r'% of hate crimes = $\frac{hate crimes count}{population}*100$',
            xy=(575, 300), xycoords='figure points')
legend = ax.legend(loc='upper right', shadow=True)

# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('#E0E0E0')
plt.title('Hate Crimes in Michigan')
plt.ylabel('Percentage of Hate Crimes')
plt.xlabel('Year')

plt.figure(figsize=(10,10))

plt.show()
```
